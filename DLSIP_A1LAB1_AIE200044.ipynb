```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler

# Load the data into a pandas DataFrame
df = pd.read_csv('Custom_CNN_Features.csv')

# Separate the target class from the feature data
X = df.drop('Class Label', axis=1)
y = df['Class Label']

# Apply the RandomOverSampler to balance the data
ros = RandomOverSampler()
X_resampled, y_resampled = ros.fit_resample(X, y)

# Combine the balanced data back into a DataFrame
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['Class Label'])], axis=1)

# Save the balanced data to a new file
df_resampled.to_csv('balanced_data.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
df = pd.read_csv('Custom_CNN_Features.csv')

# Separate the target class from the feature data
data = df.drop('Class Label', axis=1)
data = data.drop('Filename', axis = 1)
labels = df['Class Label']

train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.25)
```

```python
import seaborn as sns
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_data, train_labels)

predictions = knn.predict(test_data)
accuracy = accuracy_score(test_labels, predictions)
precision = precision_score(test_labels, predictions,average='macro')
recall = recall_score(test_labels, predictions,average='macro')
f1 = f1_score(test_labels, predictions,average='macro')


print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

dtc = DecisionTreeClassifier()
dtc.fit(train_data, train_labels)

predictions = dtc.predict(test_data)
accuracy = accuracy_score(test_labels, predictions)
precision = precision_score(test_labels, predictions,average='macro')
recall = recall_score(test_labels, predictions,average='macro')
f1 = f1_score(test_labels, predictions,average='macro')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```
